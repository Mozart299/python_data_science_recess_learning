```python
#Creating dataframe

import random

#generate same random numbers every time, number used doesnt matter
#The seed() method is used to initialize the random number generator
#random.seed(1)

#creating list of names
names = ["Jess","Jordan","Sandy","Ted","Bareny","Tyler","rebecca"]

#Creating ages for the names
#random.randint returns random numbers between 18 and 35 (18 and 35 inclusive)
#looping through the length of the names giving each a random number in range
ages = [random.randint(18,35) for x in range(len(names))]

#creating dictionary 'people'
people = {
    'names': names,
    'ages': ages
}
#dict method converts dictionary into pandas data frame
df = pd.DataFrame.from_dict(people)

#this also works
print(pd.DataFrame(people))
print("\n")
print (df)
```
